```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/folder/cleaned_data_with_transaction_year.csv")

print(f"Dataset shape: {df.shape}")
df.head()
```

```python
sns.histplot(df['price_per_unit_area'], kde=True)
plt.title("Price Distribution")
plt.xlabel("Price per Unit Area")
plt.ylabel("Frequency")
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", center=0, square=True)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```

***Analysis Results***

**Price Distribution**

    Distribution type: Right-skewed
    Peak frequency: 40-45 price per unit area (~75 occurrences)
    Data concentration: Most properties priced between 20-60 per unit area
    Outliers: Few properties at 100+ price range

**Correlations**

    Strong Positive:
        transaction_date & transaction_year: 0.82
        num_convenience_stores & price_per_unit_area: 0.57
    Strong Negative:
        distance_to_mrt & longitude: -0.81
        price_per_unit_area & distance_to_mrt: -0.67
        distance_to_mrt & latitude: -0.59
