```python
# week2_outliers_duplicates.ipynb

# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned data
df = pd.read_csv("../data/cleaned_data.csv")

df.columns = ['transaction_date', 'house_age', 'distance_to_mrt', 'num_convenience_stores', 'latitude', 'longitude', 'price']

# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Drop duplicates if any
df = df.drop_duplicates()

# Boxplot to visualize outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['house_age', 'distance_to_mrt', 'price']])
plt.title("Boxplot for Outlier Detection")
plt.grid()
plt.show()

# Optional: Filter out extreme price outliers (top 1%)
threshold = df['price'].quantile(0.99)
df_filtered = df[df['price'] <= threshold]

print("Shape after removing top 1% price outliers:", df_filtered.shape)

# Save cleaned data
# df_filtered.to_csv("data/cleaned_data.csv", index=False)
```
