```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Real estate.csv')
```

```python
df.rename(columns={
    'X1 transaction date': 'transaction_date',
    'X2 house age': 'house_age',
    'X3 distance to the nearest MRT station': 'distance',
    'X4 number of convenience stores': 'convenience_stores',
    'Y house price of unit area': 'price'
}, inplace=True)
```

```python
df.rename(columns={
    'X5 latitude': 'latitude',
    'X6 longitude': 'longitude'
}, inplace=True)

df.head()
```

```python
# Correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Correlation of each column with 'price'
print(correlation_matrix['price'].sort_values(ascending=False))
```

```python
df.drop(columns=['distance', 'house_age', 'convenience_stores'], inplace=True)
```

```python
df.head()
```

```python
#1. Histogram of Price (target variable)
import matplotlib.pyplot as plt

df['price'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

```python
import seaborn as sns

sns.scatterplot(x='latitude', y='price', data=df)
plt.title('Latitude vs Price')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x='longitude', y='price', data=df)
plt.title('Longitude vs Price')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x='transaction_date', y='price', data=df)
plt.title('Transaction Date vs Price')
plt.grid(True)
plt.show()
```

```python
df.groupby('transaction_date')['price'].mean().plot(kind='bar')
plt.title("Average Price by Transaction Year")
plt.xlabel("Year")
plt.ylabel("Avg Price")
plt.show()
```
